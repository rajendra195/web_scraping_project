```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.worldometers.info/gdp/nepal-gdp/"

html = requests.get(url).content

soup = BeautifulSoup(html, "html.parser")

soup
```

```python
table = soup.find_all("table")[1]
title = table.find_all("th")
table_title = [item.text for item in title]
table_title
```

```python
import pandas as pd
df = pd.DataFrame(columns=table_title)
df
```

```python
all_table_row = table.find_all("tr")
year = []
gdp_nominal = []
gdp_real = []
gdp_change = []
gdp_per_capita = []
pop_change = []
population = []
for row in all_table_row[1:]:
    row_data = row.find_all("td")
    data = [item.text for item in row_data]
    year.append(data[0])
    gdp_nominal.append(data[1])
    gdp_real.append(data[2])
    gdp_change.append(data[3])
    gdp_per_capita.append(data[4])
    pop_change.append(data[5])
    population.append(data[6])
```

```python
data = {
    "Year": year,
    "GDP Nominal (Current USD)": gdp_nominal,
    "GDP Real (Inflation adj.)": gdp_real,
    "GDP change": gdp_change,
    "GDP per capita": gdp_per_capita,
    "Pop change": pop_change,
    "Population": population,
}
```

```python
data
```

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
df
```

```python
data_in_db = df.to_records(index=False)
data_in_db
```

```python
import sqlite3

connection = sqlite3.connect("data97.db")

cur = connection.cursor()

query = """create table gdp1(year varchar[20], nominal varchar[20], real varchar[20], change varchar[20], percapita varchar[20], popchange varchar[20], population varchar[20])"""
try:
    cur.execute(query)
except Exception as e:
    print(e)
```

```python
query = """insert into gdp1(year, nominal, real, change, percapita, popchange, population)
            values(?, ?, ?, ?, ?, ?, ?)"""
cur.execute(query, data_in_db)
```
