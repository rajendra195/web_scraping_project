```python
import requests
```

```python
url = "https://www.worldometers.info/gdp/nepal-gdp/"
html = requests.get(url).content
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
table = soup.find_all("table")[1]
title = table.find_all("th")
table_title = [item.text for item in title]
table_title
```

```python
import pandas as pd
df = pd.DataFrame(columns=table_title)
df
```

```python
all_table_row = table.find_all("tr")
for row in all_table_row[1:]:
    row_data = row.find_all("td")
    data = [item.text for item in row_data]
    length = len(df)
    df.loc[length] = data
```

```python
df
```

```python
data_csv = df.to_csv("data.csv", index=False)
```

```python
data_to_db = df.to_records(index=False)
```

```python
import sqlite3
```

```python
con = sqlite3.connect("data.db")
cur = con.cursor()
```

```python
query = """create table gdp(year varchar[20], nominal varchar[20], real varchar[20], change varchar[20], percapita varchar[20], popchange varchar[20], population varchar[20])"""
```

```python
try:
    cur.execute(query)
except Exception as e:
    print(e)
```

```python
query = """insert into gdp(year, nominal, real, change, percapita, popchange, population)
            values(?, ?, ?, ?, ?, ?, ?)"""
cur.executemany(query, data_to_db)
```
